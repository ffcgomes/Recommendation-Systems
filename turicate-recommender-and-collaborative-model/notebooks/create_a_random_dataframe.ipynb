```python
import numpy as np
import pandas as pd
customerID = np.random.randint(0,500,size=50000)
#print(customerID)
productID = np.random.randint(0,50,size=50000)
#print(productID)
#df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
df=pd.DataFrame(list(zip(customerID, productID)),columns=['customerID','productID'])
df
```

```python
df=df.groupby(['customerID','productID']).productID.count()
df=pd.DataFrame(df)
df=df.rename({'productID':'purchase_count'},axis='columns')
#df['customerID']=df['customerID'].ffill()
df
```
